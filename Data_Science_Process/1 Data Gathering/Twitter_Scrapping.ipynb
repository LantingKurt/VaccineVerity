```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import datetime
from tqdm.notebook import tqdm
import pytz



# Creating list to append tweet data to
tweets_list2 = []

#search combination of terms separated by spaces
search_term = "Bakuna di epektib"


#Customize Dates by adding/removing

years = [
    "2018",
    "2019",
    "2020",
    "2021",
    "2022"
]

months = [
    "01",
    "02",
    "03",
    "04",
    "05",
    "06",
    "07",
    "08",
    "09",
    "10",
    "11",
    "12"
]

# years = [
#     "2022"
# ]

# months = [
#     "01",
# ]


end_of_month_days = [
    "31",
    "28",
    "31",
    "30",
    "31",
    "30",
    "31",
    "31",
    "30",
    "31",
    "30",
    "31"
]


#Setting Variables for Timestamp Column
now_utc = datetime.datetime.now(datetime.timezone.utc)
timezone = pytz.timezone('Asia/Shanghai')
now_gmt8 = now_utc.astimezone(timezone)
#use now_gmt8.strftime('%m-%d-%Y %H:%M:%S')

# Using TwitterSearchScraper to scrape data and append tweets to list
for year in tqdm(years):
    for month in tqdm(range(len(months))):
        print(f"Starting with {year}-{months[month]}-01", end = '')
        for i, tweet in enumerate(
                sntwitter.TwitterSearchScraper(f'{search_term} since:{year}-{months[month]}-01 until:{year}-{months[month]}-{end_of_month_days[month]}').get_items()):
            if i >= 300:
                break
            tweets_list2.append([
                now_gmt8.strftime('%d/%m/%Y %H:%M:%S'),
                tweet.url,
                '@'+tweet.user.username,
                tweet.user.displayname,
                tweet.user.renderedDescription,
                tweet.user.created,
                tweet.user.friendsCount,
                tweet.user.followersCount,
                tweet.user.location,
                tweet.renderedContent,
                tweet.date,
                tweet.likeCount,
                tweet.replyCount,
                tweet.retweetCount,
                tweet.quoteCount,
                tweet.viewCount])
        print(f" DONE!")

# Creating a dataframe (table) from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=[
    "Timestamp",
    "Tweet URL",
    "Account Handle",
    "Account Name",
    "Account Bio",
    "Joined",
    "Following",
    "Followers",
    "Location",
    "Tweet",
    "Date Posted",
    "Likes",
    "Replies",
    "Retweets",
    "Quote Tweets",
    "Views"])

# Converting columns with date of time zone from UTC to GMT+8 and reformatting date to match gsheet dataset
tweets_df2['Joined'] = tweets_df2['Joined'].dt.tz_convert('Etc/GMT-8').dt.strftime('%m/%Y')
tweets_df2['Date Posted'] = tweets_df2['Date Posted'].dt.tz_convert('Etc/GMT-8').dt.strftime('%m/%d/%Y %H:%M:%S')



##NOTE: timezone set to GMT-8 because apparently pandas understands opposite signs :))
#(hayop na bug na to napuyat ako dito HAHAHHA)

#Output to a csv file
tweets_df2.to_csv(f'term_{search_term} date_{datetime.date.today()} time_{now_gmt8.strftime("%H:%M:%S")} .csv')

print("DONE!")
```
